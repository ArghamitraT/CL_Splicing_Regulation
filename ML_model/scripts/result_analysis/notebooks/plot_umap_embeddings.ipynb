```python
# --- Define color palette ---

def plot_Umap_brain(df, save=True, legend=["Up-regulated", "Down-regulated"]):
   
    import time
    from plotnine import (
        ggplot, aes, geom_point, geom_density_2d, labs, theme_bw, theme,
        element_text, element_blank, scale_color_manual, coord_fixed, coord_cartesian
    )

    palette = {"High": "#377eb8", "Low": "#e41a1c"}
    labels = {"High": legend[0], "Low": legend[1]}


    p = (
        ggplot(df, aes(x="x", y="y", color="status"))
        + geom_point(size=1.8, alpha=0.65)
        # + geom_density_2d(color="gray", alpha=0.3, size=0.3)  # optional contour
        + scale_color_manual(values=palette, labels=labels)
        + coord_fixed()
        + labs(x="UMAP-1", y="UMAP-2", title="Tissue-wise ρ(Δψ)", color=None)
        + theme_bw()
        + coord_cartesian(ylim=(-1, 8))  # ✅ preferred approach
        + theme(
            figure_size=(4, 4),
            legend_position=(0.6, 0.97),       # top-right corner
            legend_justification=(1, 1),
                legend_background=element_blank(),
                legend_title=element_blank(),
                legend_text=element_text(size=15),
            axis_text_x=element_text(size=15, ha="center", va="top", color="#222222"),
            axis_text_y=element_text(size=15, color="#222222"),
            axis_title_x=element_text(margin={'t': 8}, size=18, weight="bold"),
            axis_title_y=element_text(size=20, weight="bold"),
            plot_title=element_text(size=15, weight="bold", ha="center"),
            panel_border=element_blank(),
            panel_grid_major_y=element_text(color="#dddddd"),
            panel_grid_minor_y=element_blank(),
            panel_grid_major_x=element_blank(),
            panel_grid_minor_x=element_blank(),
            axis_line_x=element_text(color="black"),
            axis_line_y=element_text(color="black"),
        )
    )


    p.show()

    out_path = f"/gpfs/commons/home/atalukder/Contrastive_Learning/files/RECOMB_26/figures/fig_TissueWise_ASCOT_brainUmap_{time.strftime('%Y%m%d_%H%M%S')}"

    if save:
        # --- Draw and save with consistent size ---
        fig = p.draw()

        # Force the figure to match notebook display ratio
        fig.set_size_inches(4, 4)  # same as theme figure_size

        formats = ['png', 'svg', 'pdf', 'eps']
        for fmt in formats:
            save_path = f"{out_path}.{fmt}"
            fig.savefig(
            save_path,
            dpi=300,
            bbox_inches="tight",
            facecolor="white"
            )
            print(f"✅ Saved: {save_path}")



        print("✅ Saved: fig_spearman_SOTA_vs_CLADE_consistent.png (same as notebook view)")
```

```python
def plot_Umap_2exon(df, save=True, legend=["METAP1, Exon 8", "CD37, Exon 5"]):
    from plotnine import (
    ggplot, aes, geom_point, geom_density_2d, labs, theme_bw, theme,
    element_text, element_blank, scale_color_manual, coord_fixed, expand_limits,coord_cartesian
)
    import time

    palette = {"High": "#7570b3", "Low": "#e6ab02"}
    labels = {"High": legend[0], "Low": legend[1]}

    p = (
        ggplot(df, aes(x="x", y="y"))
        # Smooth contour outlines — `geom_density_2d()` works on all versions
        + geom_density_2d(aes(color="status"), size=0.65, alpha=0.4)
        + geom_point(aes(color="status"), size=1.8, alpha=0.75)
        + scale_color_manual(values=palette, labels=labels)
        + coord_fixed()
        + coord_cartesian(ylim=(1.8, 9.5))  # ✅ preferred approach
        + expand_limits(
            # INCREASE THE BUFFER HERE
            x=(df["x"].min() - 1.0, df["x"].max() + 1.0), # Increased buffer
            y=(df["y"].min() - 1.0, df["y"].max() + 1.0)  # Increased buffer
        )
        + labs(x="UMAP-1", y="UMAP-2", color=None)
        + theme_bw()
        + theme(
            figure_size=(4, 4),
            legend_position=(0.7, 0.96),       # top-right corner
            legend_justification=(1, 1),
                legend_background=element_blank(),
                legend_title=element_blank(),
                legend_text=element_text(size=15),
            axis_text_x=element_text(size=15, ha="center", va="top", color="#222222"),
            axis_text_y=element_text(size=15, color="#222222"),
            axis_title_x=element_text(margin={'t': 8}, size=18, weight="bold"),
            axis_title_y=element_text(size=20, weight="bold"),
            plot_title=element_text(size=15, weight="bold", ha="center"),
            panel_border=element_blank(),
            panel_grid_major_y=element_text(color="#dddddd"),
            panel_grid_minor_y=element_blank(),
            panel_grid_major_x=element_blank(),
            panel_grid_minor_x=element_blank(),
            axis_line_x=element_text(color="black"),
            axis_line_y=element_text(color="black"),
        )
    )

    p.show()

    out_path = f"/gpfs/commons/home/atalukder/Contrastive_Learning/files/RECOMB_26/figures/fig_TissueWise_ASCOT_2exon_{time.strftime('%Y%m%d_%H%M%S')}"

    if save:
        # --- Draw and save with consistent size ---
        fig = p.draw()

        # Force the figure to match notebook display ratio
        fig.set_size_inches(4, 4)  # same as theme figure_size

        formats = ['png', 'svg', 'pdf', 'eps']
        for fmt in formats:
            save_path = f"{out_path}.{fmt}"
            fig.savefig(
            save_path,
            dpi=300,
            bbox_inches="tight",
            facecolor="white"
            )
            print(f"✅ Saved: {save_path}")



        print("✅ Saved: fig_spearman_SOTA_vs_CLADE_consistent.png (same as notebook view)")
```

```python
def plot_Umap_4exon(df, save=True, legend=["TIAM2, Ex 23", "FUNDC1, Ex 2", "ASCC1, Ex 8", "KCNU1, Ex 25"]):
    from plotnine import (
        ggplot, aes, geom_point, geom_density_2d, labs, theme_bw, theme,
        element_text, element_blank, scale_color_manual,
        coord_fixed, expand_limits, coord_cartesian
    )
    import time

    # === Expect df columns: ['exon_name', 'species', 'group', 'x', 'y'] ===
    # Rename group → status for consistency
    if "status" not in df.columns and "group" in df.columns:
        df = df.rename(columns={"group": "status"})

    # === Define palette and legend ===
    palette = {
        "E1": "#7570b3",   # purple
        "E2": "#e6ab02",   # yellow
        "E3": "#d95f02",   # orange
        "E4": "#1b9e77"    # teal
    }

    # Map unique exon IDs automatically if present
    exon_labels = df["exon_name"].unique()
    if len(exon_labels) == 4:
        mapping = {exon_labels[i]: f"E{i+1}" for i in range(4)}
        df["status"] = df["exon_name"].map(mapping)
    else:
        print(f"⚠️ Found {len(exon_labels)} unique exons — expected 4.")

    labels = {f"E{i+1}": legend[i] for i in range(4)}

    # === Build plot ===
    p = (
        ggplot(df, aes(x="x", y="y"))
        + geom_density_2d(aes(color="status"), size=0.65, alpha=0.4)
        + geom_point(aes(color="status"), size=1.8, alpha=0.75)
        + scale_color_manual(values=palette, labels=labels)
        + coord_fixed()
        + coord_cartesian(xlim=(0, 20), ylim=(5, 15.5))
        + expand_limits(
            x=(df["x"].min() - 1.0, df["x"].max() + 1.0),
            y=(df["y"].min() - 1.0, df["y"].max() + 1.0)
        )
        + labs(
            x="UMAP-1",
            y="UMAP-2",
            color=None,
            title="Exon Homologs"
        )
        + theme_bw()
        + theme(
            figure_size=(5, 5),
             legend_position=(0.95, 0.96),
            # legend_position="none",
            legend_justification=(1, 1),
            legend_background=element_blank(),
            legend_title=element_blank(),
            legend_text=element_text(size=13),
            axis_text_x=element_text(size=15, ha="center", va="top", color="#222222"),
            axis_text_y=element_text(size=15, color="#222222"),
            axis_title_x=element_text(margin={'t': 8}, size=18, weight="bold"),
            axis_title_y=element_text(size=18, weight="bold"),
            plot_title=element_text(size=18, weight="bold", ha="center"),
            panel_border=element_blank(),
            panel_grid_major_y=element_text(color="#dddddd"),
            panel_grid_minor_y=element_blank(),
            panel_grid_major_x=element_blank(),
            panel_grid_minor_x=element_blank(),
            axis_line_x=element_text(color="black"),
            axis_line_y=element_text(color="black"),
        )
    )

    p.show()

    # === Save ===
    out_path = f"/gpfs/commons/home/atalukder/Contrastive_Learning/files/RECOMB_26/figures/fig_TissueWise_ASCOT_4exon_{time.strftime('%Y%m%d_%H%M%S')}"
    if save:
        fig = p.draw()
        fig.set_size_inches(4, 4)
        for fmt in ["png", "svg", "pdf", "eps"]:
            save_path = f"{out_path}.{fmt}"
            fig.savefig(save_path, dpi=300, bbox_inches="tight", facecolor="white")
            print(f"✅ Saved: {save_path}")
```
